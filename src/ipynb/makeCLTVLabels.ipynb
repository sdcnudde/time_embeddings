##### Get percentiles and make CLTV classes

```python
customer_labels = pandas.read_table('~/Downloads/customer_labels.csv', header=None, names=['cid', 'gender', 'cltv', 'churn'])
```

```python
labels_cltv = customer_labels[['cid', 'cltv']]
```

```python
labels_cltv
```

```python
cltv_ids = labels_cltv["cid"].values
cltv_values = labels_cltv["cltv"].values
```

```python
sorted_cltv = np.sort(cltv_values)
```

```python
print "10th percentile: " + str(np.percentile(sorted_cltv, 10))
print "20th percentile: " + str(np.percentile(sorted_cltv, 20))
print "30th percentile: " + str(np.percentile(sorted_cltv, 30))
print "40th percentile: " + str(np.percentile(sorted_cltv, 40))
print "50th percentile: " + str(np.percentile(sorted_cltv, 50))
print "60th percentile: " + str(np.percentile(sorted_cltv, 60))
print "70th percentile: " + str(np.percentile(sorted_cltv, 70))
print "80th percentile: " + str(np.percentile(sorted_cltv, 80))
print "90th percentile: " + str(np.percentile(sorted_cltv, 90))
print "100th percentile: " + str(np.percentile(sorted_cltv, 100))
```

```python
labels_cltv_list = np.zeros(len(cltv_values))
for i in range(0, len(cltv_values)):
    if cltv_values[i] == 0:
        labels_cltv_list[i] = 0
    if (cltv_values[i] > 0) & (cltv_values[i] < 19.5):
        labels_cltv_list[i] = 1
    if (cltv_values[i] >= 19.5) & (cltv_values[i] < 49.5):
        labels_cltv_list[i] = 2
    if (cltv_values[i] >= 49.5) & (cltv_values[i] < 93.99):
        labels_cltv_list[i] = 3
    if (cltv_values[i] >= 93.99) & (cltv_values[i] < 169.87):
        labels_cltv_list[i] = 4
    if (cltv_values[i] >= 169.87) & (cltv_values[i] < 336.79):
        labels_cltv_list[i] = 5
    if (cltv_values[i] >= 336.79) & (cltv_values[i] <= 193820):
        labels_cltv_list[i] = 6
```

```python
np.savetxt('labels_cltv.txt', np.transpose(np.vstack((cltv_ids, labels_cltv_list)).astype(int)), fmt='%i')
```
